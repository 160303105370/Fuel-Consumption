# STEP#4: CREATE TESTING AND TRAINING DATASET

```python
from sklearn.model_selection import train_test_split
```

```python
X=fuelEconomy_df['Horse Power']
y=fuelEconomy_df['Fuel Economy (MPG)']
```

```python
X
```

```python
y
```

```python
X = np.asarray(X)
X = np.reshape(X, (-1, 1))
y = np.asarray(y)
y = np.reshape(y, (-1, 1))
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=52,test_size=.25)
```

```python
X_train
```

```python
y_train
```

# STEP#5: TRAIN THE MODEL

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression(fit_intercept=True)
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coefficient(m)', regressor.coef_)
print('Linear Model Coefficient(b)', regressor.intercept_)
```

# STEP#6: TEST THE MODEL 

```python
y_predict=regressor.predict(X_test)
y_predict
```

```python
y_test
```

```python
plt.scatter(X_train,y_train,color='gray')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.xlabel('Horse power (HP)')
plt.ylabel('MPG')
plt.title('HP Vs. MPG(Training set)')
```

```python
plt.scatter(X_test,y_test,color='gray')
plt.plot(X_test,regressor.predict(X_test),color='blue')
plt.xlabel('Horse power (HP)')
plt.ylabel('MPG')
plt.title('HP Vs. MPG(Training set)')
```

```python
# Predicting Fuel economy(MPG) when the Horse power of the vehicle is 150

HP=150
HP = np.asarray(HP)
HP = np.reshape(HP, (-1, 1))
MPG=regressor.predict(HP)
MPG
```
